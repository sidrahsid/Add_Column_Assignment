```python
import pandas as pd
data = pd.read_csv("Pokemon.csv")
data.info()
data.isnull().sum()
```

```python
#Compute mean of Speed and HP
speed_mean = data["Speed"].mean()
hp_mean = data["HP"].mean()
print(speed_mean)
print(hp_mean)
def set_speed(val):
    if val < speed_mean:
        return "Low_Speed"
    elif val == speed_mean:
        return "Neutral_Speed"
    else:
        return "High_Speed"
def set_hp(val):
    if val < hp_mean:
        return "Low_Hp"
    elif val == hp_mean:
        return "Neutral_Hp"
    else:
        return "High_Hp"
```

```python
data["Speed_high_low"] = data["Speed"].apply(set_speed)
data["HP_high_low"] = data["HP"].apply(set_hp)
data
```
